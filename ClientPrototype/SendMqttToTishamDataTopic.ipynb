```python
import csv
import json
import boto3
import time
import os
def sendMQTTmessage(row):
    iot = boto3.client('iot-data');
    energyReading = {}
    energyReading['power'] = row[0]
    energyReading['time'] = row[1]
    print (energyReading)
    data = json.dumps(energyReading)
    response = iot.publish(
                     topic='/clientPrototype/data/Tisham',
                     payload=data
                )
    time.sleep(1)
filepath = os.getcwd()+'/powerAndTimes.csv'
with open(filepath,"r") as f:
    csvReader = csv.reader(f,delimiter = ",")
    for row in csvReader:
        sendMQTTmessage(row)
```
